```python
import os
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

path=r"C:\Users\amany\OneDrive\Desktop\DV PROJECT"
print("Files to direcotry")
files=os.listdir(path)
print(files[:2])

def load_CSV(filename):
    return pd.read_csv(filename)
deliveries=pd.read_csv(r"C:\Users\amany\OneDrive\Desktop\DV PROJECT\deliveries.csv")
matches=pd.read_csv(r"C:\Users\amany\OneDrive\Desktop\DV PROJECT\matches.csv")
print(deliveries.head())
print(deliveries.info())
print(matches.head())
print(matches.info())
print(f"Question 1: Matches won by maximum margin of runs")
max_runs_index=matches['win_by_runs'].idxmax()
matches.loc[max_runs_index]
print(max_runs_index)
print(f"Question 2: Matches won by maximum wickets")
max_wickets_index=matches['win_by_wickets'].idxmax()
matches.loc[max_wickets_index]
print(max_wickets_index)
won_by_runs = matches[matches['win_by_runs'] > 0]
print(f"Question 3: Matches won by minimum margin of runs ")
min_runs_index=matches['win_by_runs'].idxmin()
matches.loc[min_runs_index]
print(min_runs_index)
won_by_wickets = matches[matches['win_by_wickets'] > 0]
print(f"Question 4: Matches won by minimum wickets")
min_wickets_index=matches['win_by_wickets'].idxmin()
matches.loc[min_wickets_index]
print(min_wickets_index)
print(f"Question 5: Matches won Where D/L method was & wasn't applied")
matches['dl_applied'].value_counts()
print(matches['dl_applied'].value_counts())
print(f"Question 6:No. of matches held per city")
city_count=matches['city'].value_counts()
print(city_count)
plt.figure(figsize=(15,6))
sns.barplot(x=city_count.index,y=city_count.values)
plt.title("NUMBER OF MATCHES HELD IN EACH CITY")
plt.xticks(rotation=90)
plt.show()
##### Insights apart from already given to find
print(f"The insights we can find from the graph and data is that:\n Highest number of matches have been held in MUMBAI while Lowest number of matches have been held in BLOEMFONTEIN.\n There is sudden incline in number of matches as we move from city DURBAN to PUNE and so on.\n There are 5 cities who have equal number match hosting which is 7 by each.")
print("Question 7: Number of matches won by each team")
team_wins = matches['winner'].value_counts()
team_wins_df = team_wins.reset_index()
team_wins_df.columns = ['Team', 'Number of Wins']

plt.figure(figsize=(15, 6))
sns.barplot(x='Team', y='Number of Wins', data=team_wins_df, palette='viridis')
plt.title('Number of Matches Won by Each Team')
plt.xticks(rotation=90)
plt.xlabel('Team')
plt.ylabel('Number')
plt.show()
print(f"The insights we can find from the graph and data is that :\n Highest matches are won by Mumbai Indians.\n The lowest win count is of RISING PUNE SUPERGAINTS.\n There is a sudden incline with winning rate from DECCAN CHARGERS & between Gujarat Lions.\n while Delhi capitals and Rising Pune Supergaint.")
print(f'Question 8: Number of matches held every season')
matches_per_season = matches.groupby('season').size()
print(matches_per_season)
matches_per_season = matches.groupby('season').size()
matches_per_season_df = matches_per_season.reset_index()
matches_per_season_df.columns = ['Season', 'Number of Matches']
plt.figure(figsize=(12, 6))
sns.barplot(x='Season', y='Number of Matches', data=matches_per_season_df, palette='pastel')
plt.title('Number of Matches Held Each Season')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45)
plt.show()
print(f"The insights of the data are as follow:\n As we can see the highest number of matches have been played in season 2013. \n And the Lowest number of matches have been played in year 2009 as just in second sesason ipl. \n And in overall data matches per seasons avg matches rotates in between 60-62 matches ")
print("Question 9: TOP 10 PLAYERS BASED ON NUMBER OF MOM AWARDS WON")
player_of_match = matches['player_of_match']
top_10_players = player_of_match.value_counts().nlargest(10)
top_10_players_df = top_10_players.reset_index()
top_10_players_df.columns = ['Player', 'Awards']
print(top_10_players_df)
plt.figure(figsize=(12, 6))
sns.barplot(x='Player', y='Awards', data=top_10_players_df, palette='magma')
plt.title('Top 10 Players Based on Man of the Match Awards')
plt.xticks(rotation=45)
plt.xlabel('Player')
plt.ylabel('Number of Awards')
plt.show()
print(f"The insights we can get from the data and graph  is: \n The highest number of MOM awards have been achieved or taken by CH  GAYLE and Lowest have been taken by MK HUSSEY in between these TOP 10 players. \n There is an sudden incline between mom Awards of MS DHONI and other Top 2  players. \n There are 3 players with same number of MOM Awards won. ")
print("Question 10: Does winning the toss means winninng the match ")
toss_winner_wins = matches[matches['toss_winner'] == matches['winner']]
toss_winner_win_percentage = len(toss_winner_wins) / len(matches) * 100
print(f"Winning the toss results in a win {toss_winner_win_percentage:.2f}% of the time.")
matches['toss_winner_wins'] = matches['toss_winner'] == matches['winner']
toss_win_counts = matches['toss_winner_wins'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=toss_win_counts.index, y=toss_win_counts.values, palette='coolwarm')
plt.title('Win Percentage When Winning the Toss')
plt.xlabel('Did Toss Winner Win the Match?')
plt.ylabel('Number of Matches')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
print(f"The insights we can get is: \n In more than 50% cases its true that winning the means winning the match. \n The winning ratio or percentage of toss winning team is higher than toss lossing team. ")
print(f"Question 11: What was the decision taken by the captions when they won the toss")
toss_decision_counts = matches['toss_decision'].value_counts()
plt.figure(figsize=(8, 6))
sns.barplot(x=toss_decision_counts.index, y=toss_decision_counts.values, palette='Set2')
plt.title('Toss Decision by Captains')
plt.xlabel('Toss Decision')
plt.ylabel('Number of Matches')
plt.show()
print(f"The insights we can get from the data & graph is:\n Decision of choosing to feild is very much higher than choosing to bat. \n It also shows that players are more concerned about feilding rather than batting.")
print(f"Question 12: No. of matches where D/L  methods where applied every season")
dl_applied_per_season = matches[matches['dl_applied'] == 1].groupby('season').size()
dl_applied_per_season_df = dl_applied_per_season.reset_index()
dl_applied_per_season_df.columns = ['Season', 'Matches with D/L Method']
print(dl_applied_per_season_df)
plt.figure(figsize=(12, 6))
sns.barplot(x='Season', y='Matches with D/L Method', data=dl_applied_per_season_df, palette='viridis')
plt.title('Number of Matches with D/L Method Applied Each Season')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.xticks(rotation=45)
plt.show()
print(f"The insigts we can get from the graph and data is: \n 2016 season has highest D/L applied while 2014 & 2017 remains lowest. \n 3 seasons have equal number of D/L count which is 3 and 2 with 1 equal count.")
print(f'Question 13: Different results for matches')
match_results = matches['result'].value_counts()
print(match_results)
plt.figure(figsize=(8, 6))
sns.barplot(x=match_results.index, y=match_results.values, palette='pastel')
plt.title('Different Match Results')
plt.xlabel('Result')
plt.ylabel('Number of Matches')
plt.show()
print(f"The insights we can get from data & graph is: \n Almost each match has normal result. \n Matches TIE are ust few and same goes for no result.")
print(f'Question 14: How many times did a team win the toss')
toss_wins = matches['toss_winner'].value_counts()
print(toss_wins)
toss_wins = matches['toss_winner'].value_counts()
plt.figure(figsize=(12, 6))
sns.barplot(x=toss_wins.index, y=toss_wins.values, palette='coolwarm')
plt.title('Number of Times Each Team Won the Toss')
plt.xlabel('Team')
plt.ylabel('Number of Toss Wins')
plt.xticks(rotation=45)
plt.show()
print(f"The insights w etaken out are : \n Highest numberv of toss are won by MUMBAI INDIANS. \n Lowest toss are won by RISING PUNE SUPERGAINTS. \n There is sudden decline in number of toss won after SUNRISERS HYDERABAD. ")
print(f"Question 15: Best venue for chasing a defending a total")
defending_matches = matches[matches['win_by_runs'] > 0]
chasing_matches = matches[matches['win_by_wickets'] > 0]
defending_wins_venue = defending_matches['venue'].value_counts()
chasing_wins_venue = chasing_matches['venue'].value_counts()
venue_results = pd.DataFrame({
    'Defending Wins': defending_wins_venue,
    'Chasing Wins': chasing_wins_venue
}).fillna(0)
top_venues = matches['venue'].value_counts().head(10).index
venue_results = venue_results.loc[top_venues]
print(venue_results)
plt.figure(figsize=(12,6))
venue_results.plot(kind='bar')
plt.title("Best Venues for Defending vs Chasing")
plt.xlabel("Venue")
plt.ylabel("Number of Wins")
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
print(f"The insights we can get from this is : \n Eden Garden has highest Chasing wins while Maharastra cricket Asssociation staduim has how  chasing win rate. \n Wankhede Stadium has Highest Defending wins and Maharastra Cricket assocaition has lowest defendinng wins. ")
print(f"Question 16: Best Defending and Chasing Teams")
defending_wins_team = matches[matches['win_by_runs'] > 0]['winner'].value_counts()
chasing_wins_team = matches[matches['win_by_wickets'] > 0]['winner'].value_counts()
print(defending_wins_team)
print(chasing_wins_team)
plt.figure(figsize=(12,6))
sns.barplot(x=defending_wins_team.values, y=defending_wins_team.index)
plt.title("Best Defending Teams")
plt.xlabel("Wins")
plt.ylabel("Team")
plt.show()
print(f"Question 17: No of Matches Played in every stadium")
stadium_counts = matches['venue'].value_counts()
plt.figure(figsize=(10, 8))
sns.barplot(x=stadium_counts.values, y=stadium_counts.index)
plt.title('Total Matches Played per Stadium')
plt.show()
```
